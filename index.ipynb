### Step 4: Run a simple regression in `statsmodels` with TV as a predictor

```python
# import libraries
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('seaborn')


# build the formula 
f = "sales~TV"
# create a fitted model in one line
model = ols(formula=f, data=data).fit()
```

### Step 5: Get regression diagnostics summary


```python
model.summary()
```

### Record your observations on "Goodness of fit"


Note here that the coefficients represent associations, not causations


```python
# The R-squared value could be higher, it is negatively skewed, but the kurtosis is okay. 
```

### Step 6:  Draw a prediction line with data points omn a scatter plot for X (TV) and Y (Sales)

Hint: We can use `model.predict()` functions to predict the start and end point of of regression line for the minimum and maximum values in the 'TV' variable. 

```python
# create a DataFrame with the minimum and maximum values of TV

predmax = max(model.predict())
predmin = min(model.predict())
d = {'max': (data.TV.max(),predmax), 'min': (data.TV.min(), predmin)}
print(d)
# first, plot the observed data and the least squares line
plt.figure()
plt.scatter(x = data.TV, y = data.sales)
regline = []
for x in data.TV:
    regline.append(7.0326 + x*0.0475)
plt.plot(data.TV, regline, color= "REd")
```

### Step 7: Visualize the error term for variance and heteroscedasticity

```python
fig = plt.figure(figsize=(15,8))
fig = sm.graphics.plot_regress_exog(model, "TV", fig=fig)
```

```python
# Record Your observations on residuals
# The data is linearly correlated, but shows some heteroskedasticity, 
```

### Next, repeat above for radio and go through the same process, recording your observations

```python
# build the formula 
f = "sales~radio"
# create a fitted model in one line
model = ols(formula=f, data=data).fit()
predmax = max(model.predict())
predmin = min(model.predict())
d = {'max': (data.radio.max(),predmax), 'min': (data.radio.min(), predmin)}
print(d)
```

```python
model.summary()
plt.figure()
plt.scatter(x = data.radio, y = data.sales)
regline = []
for x in data.radio:
    regline.append(9.3116 + x*0.2025)
plt.plot(data.radio, regline, color= "Red")
# multiple R-squared is way lower, it is highly skewed negatively, and the kurtosis is high. The fit is not as good.
```

```python
fig = plt.figure(figsize=(15,8))
fig = sm.graphics.plot_regress_exog(model, "radio", fig=fig)
```

```python
# There is less linear correlation and more heteroskedasticity.
```
